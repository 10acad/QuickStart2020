Github is a great page for portfolio with codes and pages in one place to show our data science work. 

In this lesson we will discuss how to setup a data science portfolio website for FREE using GitHub pages.

To getstarted with, you’ll need to have familiary with some of GitHub and [Git](https://git-scm.com/book/en/v2/Getting-Started-What-is-Git%3F).Git is a [version control system](https://git-scm.com/book/en/v1/Getting-Started-About-Version-Control) that allows you to track changes to your code, and collaborate with others.

[GitHub](https://github.com/features) is the platform where you can host your code work with others. 
If you are note familiar with Git or GitHub, here’s some free resources to get you started:
*   [Git Videos](https://git-scm.com/videos)
*   [Git Tutorial: Learn Git Basics For Free | Codecademy](https://www.codecademy.com/learn/learn-git)
*   [Git & GitHub Crash Course For Beginners](https://www.youtube.com/watch?v=SWYqp7iY_Tc)

If you don’t already have a GitHub profile, take a minute to [sign up for free](https://github.com/join?source=header-home). You’ll just need a valid email address to get going.

GitHub Pages are public webpages hosted and easily published through GitHub. After you login it’s time to clone the data science portfolio sample created for this lesson purpose to your GitHub. 

For this lesson purpose, we have created a basic data science portfolio template built on top of the [jekyll-theme-cayman](https://github.com/pages-themes/cayman). For advanced customization, its highly recommended to check out the awesome documentation from the theme link.

Go ahead and navigate to the [10acadstud](https://github.com/10acadstud/10acadstud.github.io) GitHub Profile.

Once you’re there, go ahead and fork the repository.

You’ll see the Fork button in the top right corner.

Click on that button and edit index.md file to your custom page.






The quickest way to get up and running is by using the Jekyll Theme Chooser to load a pre-made theme. You can then modify your GitHub Pages’ content and style remotely via the web or locally on your computer. To create yor own you can follow an awesome 10 minute tutorial on creating github pages by guthub. [Getting Started with GitHub Pages](https://guides.github.com/features/pages/)
